# Limpieza de Movimientos

```python
# CELDA 1 (Movimientos)
import pandas as pd
from pathlib import Path
import glob

carpeta = r'c:\Users\rdiezl\Desktop\proyecto\planificacion\Datos\Movimientos'
archivos = sorted(glob.glob(f'{carpeta}/*.xls') + glob.glob(f'{carpeta}/*.xlsx'))

dfs = []
for archivo in archivos:
    # La fila 2 tiene los nombres de columnas
    df = pd.read_excel(archivo, header=1, dtype={'Paleta': 'string'})

    # Limpiar Paleta (quitar solo el sufijo .0 si aparece)
    if 'Paleta' in df.columns:
        df['Paleta'] = (
            df['Paleta']
            .astype('string')
            .str.strip()
            .fillna('')
            .str.replace(r'\.0+$', '', regex=True)
        )

    # Quitar última fila (fila combinada/resumen)
    if len(df) > 0:
        df = df.iloc[:-1].copy()

    dfs.append(df)

movimientos = pd.concat(dfs, ignore_index=True)

print(f"Total de archivos leídos: {len(archivos)}")
print(f"Forma del dataframe: {movimientos.shape}")
movimientos.head()
```

```python
# CELDA 2 (Movimientos pedidos)
carpeta_pedidos = r'c:\Users\rdiezl\Desktop\proyecto\planificacion\Datos\Movimientos pedidos'
archivos_pedidos = sorted(glob.glob(f'{carpeta_pedidos}/*.xls') + glob.glob(f'{carpeta_pedidos}/*.xlsx'))

dfs_pedidos = []
for archivo in archivos_pedidos:
    # La fila 2 tiene los nombres de columnas
    df = pd.read_excel(archivo, header=1, dtype={'Paleta': 'string'})

    # Limpiar Paleta (quitar solo el sufijo .0 si aparece)
    if 'Paleta' in df.columns:
        df['Paleta'] = (
            df['Paleta']
            .astype('string')
            .str.strip()
            .fillna('')
            .str.replace(r'\.0+$', '', regex=True)
        )

    # Quitar última fila (fila combinada/resumen)
    if len(df) > 0:
        df = df.iloc[:-1].copy()

    dfs_pedidos.append(df)

movimientos_pedidos = pd.concat(dfs_pedidos, ignore_index=True)

print(f"Total de archivos leídos: {len(archivos_pedidos)}")
print(f"Forma del dataframe: {movimientos_pedidos.shape}")
movimientos_pedidos.head()
```

```python
# CELDA 5 (Ubicacion correcta + regla 301)
cols = ['Pas. Ori.', 'Col. Ori.', 'Alt. Ori.']

for c in cols:
    movimientos[c] = pd.to_numeric(movimientos[c], errors='coerce')

mask_col_301 = movimientos['Col. Ori.'].eq(301)

movimientos['Ubicacion'] = pd.NA
mask_norm = movimientos[cols].notna().all(axis=1) & ~mask_col_301

movimientos.loc[mask_norm, 'Ubicacion'] = (
    movimientos.loc[mask_norm, 'Pas. Ori.'].astype('Int64').astype(str).str.zfill(3) + '-' +
    movimientos.loc[mask_norm, 'Col. Ori.'].astype('Int64').astype(str).str.zfill(3) + '-' +
    movimientos.loc[mask_norm, 'Alt. Ori.'].astype('Int64').astype(str).str.zfill(2)
)

movimientos.loc[mask_col_301, 'Ubicacion'] = '301'

movimientos[cols + ['Ubicacion']].head()
```

```python
# CELDA 6 (Paleta sin inventar valores)
import re

def _normaliza_paleta(v):
    if pd.isna(v):
        return ''
    s = str(v).strip()
    if s == '' or s.lower() in {'nan', 'none', '<na>'}:
        return ''
    if re.fullmatch(r'\d+\.0+', s):
        return s.split('.')[0]
    return s

for name, df in (('movimientos', movimientos),
                 ('movimientos_pedidos', movimientos_pedidos)):
    if 'Paleta' in df.columns:
        df['Paleta'] = df['Paleta'].map(_normaliza_paleta)
        print(f"{name} - primeros valores de Paleta:")
        print(df['Paleta'].head())
```

```python
keys = ['Fecha inicio', 'Paleta', 'Cantidad']

# normalizar tipos/formato en ambas tablas
for df in (movimientos, movimientos_pedidos):
    df['Paleta'] = (
        df['Paleta']
        .astype('string')
        .str.strip()
        .fillna('')
        .replace({'<NA>': '', 'nan': '', 'None': ''})
    )
    df['Cantidad'] = pd.to_numeric(df['Cantidad'], errors='coerce')
    df['Fecha inicio'] = pd.to_datetime(df['Fecha inicio'], errors='coerce', dayfirst=True)

    # Si quieres que compare solo fecha (ignorando hora), descomenta:
    # df['Fecha inicio'] = df['Fecha inicio'].dt.normalize()

# evitar duplicados en la derecha para que no se multipliquen filas
movimientos_pedidos_u = movimientos_pedidos.drop_duplicates(subset=keys)

# merge horizontal: conserva todas las filas de movimientos
movimientos_merge = movimientos.merge(
    movimientos_pedidos_u,
    on=keys,
    how='left',
    suffixes=('', '_pedido')
)

movimientos_merge
```

```python
import re
import pandas as pd

def normalizar_pedido_externo(v):
    if pd.isna(v):
        return ''
    s = str(v).strip()
    if s == '' or s.lower() in {'nan', 'none', '<na>'}:
        return ''

    s_up = s.upper()

    # 1) SGE/EGE/SGP + dígitos (corta sufijos tipo -01, /24, espacio, etc.)
    m = re.match(r'^(SGE|EGE|SGP)(\d+)', s_up)
    if m:
        return f"{m.group(1)}{m.group(2)}"

    # 2) E0M2506/... -> M2506
    m = re.match(r'^E0(M\d{4})', s_up)
    if m:
        return m.group(1)

    # 3) M2506/1 (o M2506-1) -> M2506
    m = re.match(r'^(M\d{4})', s_up)
    if m:
        return m.group(1)

    # 4) resto: sin cambios
    return s

# Aplicar en dataframes disponibles y en cualquier columna "Pedido externo*"
for nombre_df in ('movimientos_pedidos', 'movimientos_merge'):
    if nombre_df in globals():
        df = globals()[nombre_df]
        cols_objetivo = [c for c in df.columns if c.startswith('Pedido externo')]
        for col in cols_objetivo:
            antes = df[col].astype('string').fillna('')
            df[col] = df[col].map(normalizar_pedido_externo)
            cambios = int((antes != df[col].astype('string').fillna('')).sum())
            print(f"{nombre_df}.{col}: {cambios} valores normalizados")
```
