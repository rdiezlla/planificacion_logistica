```python
# Rehacemos el merge nuevamente para obtener los valores DESCRIPCIÓN original sin modificar
# Mantenemos df1 (Albaranes) a la LEFT y df2 (Portes) a la RIGHT para consistencia
df_merged_nuevo = pd.merge(df1, df2, on='DESCRIPCIÓN', how='left')

# Eliminamos las mismas columnas que antes
columnas_a_eliminar = ['CIF FACTURACIÓN', 'ALBARÁN OUT', 'HUECOS PALÉ', 'M2 SUELO', 'EVENTOS "A" NORMAL', 'EVENTOS "A" FESTIVO', 'EVENTOS "B" NORMAL', 'EVENTOS "B" FESTIVO', 'MENTORING 2 PERSONAS JUEVES LABORABLE', 'MENTORING 3 PERSONAS JUEVES LABORABLE', 'MENTORING 2 PERSONAS JUEVES FESTIVO', 'MENTORING 3 PERSONAS JUEVES FESTIVO', 'DISTANCIA EVENTO A ALMACÉN (KM)', 'KM EXTRA', 'HORAS EXTRA', 'LAVADO COPAS (EXCEPTO MENTORING)', 'PRESUPUESTO PARTICULAR', 'EMPRESA', 'MES', 'AÑO']
df_merged_nuevo = df_merged_nuevo.drop(columns=columnas_a_eliminar, errors='ignore')

# Reasignar el volumen al nuevo dataframe
df_merged_nuevo['Volumen'] = (df_merged_nuevo['LARGO (cm)'] * df_merged_nuevo['ANCHO (cm)'] * df_merged_nuevo['ALTO (cm)']) / 1000000

# Encontrar posición de PESO (KG) y reordenar
posicion_peso = df_merged_nuevo.columns.get_loc('PESO (KG)')
cols = df_merged_nuevo.columns.tolist()
if 'Volumen' in cols:
    cols.remove('Volumen')
cols.insert(posicion_peso + 1, 'Volumen')
df_merged_nuevo = df_merged_nuevo[cols]

# Eliminar las columnas de dimensiones
df_merged_nuevo = df_merged_nuevo.drop(columns=['LARGO (cm)', 'ANCHO (cm)', 'ALTO (cm)'], errors='ignore')

# Ahora normalizamos correctamente
def normalizar_descripcion(valor):
    if pd.isna(valor):
        return valor
    
    valor_str = str(valor).strip()
    
    # Primero quitamos 'EXPEDICIÓN E0'
    if valor_str.startswith('EXPEDICIÓN E0'):
        valor_str = valor_str.replace('EXPEDICIÓN E0', '', 1).strip()
    
    # Quitamos espacios y caracteres especiales al final
    valor_str = valor_str.split('/')[0].strip()  # Quita everything after /
    
    # Ahora aplicamos las reglas específicas
    if any(x in valor_str for x in ['SGE', 'EGE', 'SGP']):
        # Para SGE, EGE, SGP: deben tener 12 caracteres
        return valor_str[:12]
    elif 'M' in valor_str[:1]:  # Si comienza con M
        # Para M: deben tener 5 caracteres
        return valor_str[:5]
    else:
        # Para números u otros: dejar como está
        return valor_str

df_merged_nuevo['DESCRIPCIÓN'] = df_merged_nuevo['DESCRIPCIÓN'].apply(normalizar_descripcion)

print("Valores de DESCRIPCIÓN después de la nueva normalización:")
print("\nValores con SGE, EGE o SGP:")
sg_values = df_merged_nuevo[df_merged_nuevo['DESCRIPCIÓN'].str.contains('SGE|EGE|SGP', na=False, case=False)]['DESCRIPCIÓN'].unique()
for val in sg_values[:20]:
    print(f"'{val}' - Length: {len(val)}")

print("\nValores con M:")
m_values = df_merged_nuevo[df_merged_nuevo['DESCRIPCIÓN'].str.contains('^M', na=False, case=False, regex=True)]['DESCRIPCIÓN'].unique()
for val in m_values[:20]:
    print(f"'{val}' - Length: {len(val)}")

print("\nValores solo números:")
num_values = df_merged_nuevo[df_merged_nuevo['DESCRIPCIÓN'].str.match(r'^\d+$', na=False)]['DESCRIPCIÓN'].unique()
for val in num_values[:20]:
    print(f"'{val}' - Length: {len(val)}")
```

```python
# Normalización mejorada que maneja espacios y caracteres especiales
def normalizar_descripcion_v2(valor):
    if pd.isna(valor):
        return valor
    
    valor_str = str(valor).strip()
    
    # Quitar 'EXPEDICIÓN E0' del inicio
    if valor_str.startswith('EXPEDICIÓN E0'):
        valor_str = valor_str.replace('EXPEDICIÓN E0', '', 1).strip()
    
    # Quitar espacios (pueden estar en mitad del código)
    valor_sin_espacios = valor_str.replace(' ', '')
    
    # Quitar todo después de la barra (año)
    valor_sin_ano = valor_sin_espacios.split('/')[0].strip()
    
    # Ahora aplicar las reglas específicas
    if any(x in valor_sin_ano for x in ['SGE', 'EGE', 'SGP']):
        # Para SGE, EGE, SGP: deben tener exactamente 12 caracteres
        # Si tiene más de 12, truncar; si tiene menos de 12, dejar como está (pero lo ideal sería rellenar)
        return valor_sin_ano[:12] if len(valor_sin_ano) >= 12 else valor_sin_ano
    elif valor_sin_ano.startswith('M'):  
        # Para M: deben tener 5 caracteres
        return valor_sin_ano[:5] if len(valor_sin_ano) >= 5 else valor_sin_ano
    else:
        # Para números u otros: dejar como está
        return valor_sin_ano

# Aplicar la nueva normalización al dataframe
df_merged_nuevo['DESCRIPCIÓN'] = df_merged_nuevo['DESCRIPCIÓN'].apply(normalizar_descripcion_v2)

# Recalcular desde los datos originales con la nueva normalización
# Mantenemos df1 (Albaranes) a la LEFT y df2 (Portes) a la RIGHT para consistencia
df_merged_nuevo_v2 = pd.merge(df1, df2, on='DESCRIPCIÓN', how='left')
df_merged_nuevo_v2 = df_merged_nuevo_v2.drop(columns=columnas_a_eliminar, errors='ignore')
df_merged_nuevo_v2['Volumen'] = (df_merged_nuevo_v2['LARGO (cm)'] * df_merged_nuevo_v2['ANCHO (cm)'] * df_merged_nuevo_v2['ALTO (cm)']) / 1000000

posicion_peso = df_merged_nuevo_v2.columns.get_loc('PESO (KG)')
cols = df_merged_nuevo_v2.columns.tolist()
if 'Volumen' in cols:
    cols.remove('Volumen')
cols.insert(posicion_peso + 1, 'Volumen')
df_merged_nuevo_v2 = df_merged_nuevo_v2[cols]
df_merged_nuevo_v2 = df_merged_nuevo_v2.drop(columns=['LARGO (cm)', 'ANCHO (cm)', 'ALTO (cm)'], errors='ignore')

# Aplicar normalización
df_merged_nuevo_v2['DESCRIPCIÓN'] = df_merged_nuevo_v2['DESCRIPCIÓN'].apply(normalizar_descripcion_v2)

# Mostrar resultados
print("Valores de DESCRIPCIÓN después de la normalización mejorada:")
print("\nValores con SGE, EGE o SGP:")
sg_values = df_merged_nuevo_v2[df_merged_nuevo_v2['DESCRIPCIÓN'].str.contains('SGE|EGE|SGP', na=False, case=False)]['DESCRIPCIÓN'].unique()
for val in sg_values[:20]:
    print(f"'{val}' - Length: {len(val)}")

print("\nValores con M:")
m_values = df_merged_nuevo_v2[df_merged_nuevo_v2['DESCRIPCIÓN'].str.contains('^M', na=False, case=False, regex=True)]['DESCRIPCIÓN'].unique()
for val in m_values[:15]:
    print(f"'{val}' - Length: {len(val)}")

# Reemplazar df_merged_limpio con la versión correcta
df_merged_limpio = df_merged_nuevo_v2.copy()
```

```python
# Normalización final que rellena códigos SGE/EGE/SGP incompletos con ceros
def normalizar_descripcion_final(valor):
    if pd.isna(valor):
        return valor
    
    valor_str = str(valor).strip()
    
    # Quitar 'EXPEDICIÓN E0' del inicio
    if valor_str.startswith('EXPEDICIÓN E0'):
        valor_str = valor_str.replace('EXPEDICIÓN E0', '', 1).strip()
    
    # Quitar espacios (pueden estar en mitad del código)
    valor_sin_espacios = valor_str.replace(' ', '')
    
    # Quitar todo después de la barra (año)
    valor_sin_ano = valor_sin_espacios.split('/')[0].strip()
    
    # Ahora aplicar las reglas específicas
    if any(x in valor_sin_ano for x in ['SGE', 'EGE', 'SGP']):
        # Para SGE, EGE, SGP: deben tener exactamente 12 caracteres
        # Si tienen menos, rellenar con ceros a la derecha
        if len(valor_sin_ano) < 12:
            valor_sin_ano = valor_sin_ano.ljust(12, '0')
        return valor_sin_ano[:12]
    elif valor_sin_ano.startswith('M'):  
        # Para M: deben tener 5 caracteres
        if len(valor_sin_ano) < 5:
            valor_sin_ano = valor_sin_ano.ljust(5, '0')
        return valor_sin_ano[:5]
    else:
        # Para números u otros: dejar como está
        return valor_sin_ano

# Aplicar la normalización final
df_merged_limpio['DESCRIPCIÓN'] = df_merged_limpio['DESCRIPCIÓN'].apply(normalizar_descripcion_final)

print("Valores de DESCRIPCIÓN después de la normalización FINAL:")
print("\nValores con SGE, EGE o SGP:")
sg_values = df_merged_limpio[df_merged_limpio['DESCRIPCIÓN'].str.contains('SGE|EGE|SGP', na=False, case=False)]['DESCRIPCIÓN'].unique()
for val in sorted(sg_values):
    print(f"'{val}' - Length: {len(val)}")

print(f"\nTotal de valores SGE/EGE/SGP únicos con 12 caracteres: {(df_merged_limpio[df_merged_limpio['DESCRIPCIÓN'].str.contains('SGE|EGE|SGP', na=False, case=False)]['DESCRIPCIÓN'].str.len() == 12).sum()}")

print("\nValores con M:")
m_values = df_merged_limpio[df_merged_limpio['DESCRIPCIÓN'].str.contains('^M', na=False, case=False, regex=True)]['DESCRIPCIÓN'].unique()
for val in sorted(m_values)[:20]:
    print(f"'{val}' - Length: {len(val)}")

print(f"\nTotal de valores M únicos con 5 caracteres: {(df_merged_limpio[df_merged_limpio['DESCRIPCIÓN'].str.contains('^M', na=False, case=False, regex=True)]['DESCRIPCIÓN'].str.len() == 5).sum()}")
```

```python
# Crear columna 'CAJAS OUT' a partir de 'CAJAS SIN MANIPULACIÓN' y 'CAJAS CON MANIPULACIÓN'
def _cajas_out(a, b):
    if pd.isna(a) and pd.isna(b):
        return pd.NA
    if pd.isna(a):
        return b
    if pd.isna(b):
        return a
    # ambas presentes
    return a if a == b else (a if a > b else b)

df_merged_limpio['CAJAS OUT'] = df_merged_limpio['CAJAS SIN MANIPULACIÓN'].combine(
    df_merged_limpio['CAJAS CON MANIPULACIÓN'], _cajas_out)

# Convertir a numérico cuando sea posible
df_merged_limpio['CAJAS OUT'] = pd.to_numeric(df_merged_limpio['CAJAS OUT'], errors='coerce')

# Eliminar las columnas originales
df_merged_limpio = df_merged_limpio.drop(columns=['CAJAS SIN MANIPULACIÓN', 'CAJAS CON MANIPULACIÓN'], errors='ignore')

# Mostrar comprobación
df_merged_limpio[['CAJAS OUT']].head()
```
